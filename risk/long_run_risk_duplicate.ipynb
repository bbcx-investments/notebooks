```python
from scipy.stats import norm

# example normal distribution of returns
mean = 0.06
standard_deviation = 0.15
rv = norm(loc=mean, scale=standard_deviation)

# example year invested
T = 20
# simulating T returns from the normal distribution
ret = rv.rvs(T)
ret
```

```python
compounded_return = (1+ret).prod()
print('The simulated compounded return of', T, 'years is', compounded_return, 'when the returns are normally distributed with mean', mean, 'and standard deviation', standard_deviation)
```

We can simulate more compounded returns to have a simulated distribution of compounded returns.

```python
numsims = 1000  # number of simulations
ret = rv.rvs(T * numsims).reshape((T, numsims))
ret
```

```python
compounded_returns = (1 + ret).prod(axis=0)
compounded_returns
```

```python
import pandas as pd
df = pd.DataFrame(compounded_returns)
df.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```
